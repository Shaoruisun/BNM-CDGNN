```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font',family='Arial',weight= 'bold')
fig, ax = plt.subplots(figsize =(8,8),dpi = 1000)

df = pd.read_csv("history_5_8_12.csv")

train_mae = df.iloc[:, 2:3]
val_mae = df.iloc[:, 4:5]

print(val_mae.min())
training_losses = train_mae
validation_losses = val_mae

# 设置刻度字号
font_size = 20
ax.tick_params(labelsize=font_size)

plt.plot(training_losses, label="Training mae", linewidth=0.5,c='#4CAAAC')
plt.plot(validation_losses, label="Validation mae", linewidth=0.5, c='darkorange')

ax.set_xlabel('Epoch', fontsize=20,weight= 'bold')
ax.set_ylabel('Mean Absolute Error of Adsorption Energy (eV)', fontsize=20,weight= 'bold')

#ax.tick_params(axis='x', colors='black', l=35, width=6, length=15, color='black')
#ax.tick_params(axis='y', colors='black', labelsize=35, width=6, length=15, color='black')

for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(1)

plt.legend(frameon=True, fontsize=15)
plt.savefig("Ef_4_loss_curve.jpeg")
plt.show()
```
